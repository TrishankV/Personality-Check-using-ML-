As everythinh is affecting the target Y = personality we are going to apply multi-linear regression and find the accuracy 

```python
from sklearn.model_selection import train_test_split as tts 
```

```python
X = data.drop(columns=['Personality'])
y = data['Personality']
```

```python
X = X.apply(pd.to_numeric, errors='raise')  # Will raise an error if conversion fails
```

```python
X_train , X_test , y_train , y_test = tts(X , y , test_size = 0.3 , random_state = 42)
```

```python
columns_to_scale =   [
        'Time_spent_Alone',
    'Social_event_attendance',
    'Going_outside',
    'Friends_circle_size',
    'Post_frequency' ]
scaler = StandardScaler()
X_train[columns_to_scale] = scaler.fit_transform(X_train[columns_to_scale])

# Transform the same columns in the test set using the *fitted* scaler
X_test[columns_to_scale] = scaler.transform(X_test[columns_to_scale])
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(random_state=42, solver='liblinear', C=1.0)
```

```python
history = model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy Score: {accuracy*100:.4f} %")

# Confusion Matrix
# Rows are actual labels, columns are predicted labels
# [[True Negative, False Positive],
#  [False Negative, True Positive]]
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

# Classification Report (Precision, Recall, F1-score for each class)
class_report = classification_report(y_test, y_pred, target_names=lee.classes_)
print("\nClassification Report:")
print(class_report)
```

```python
feature_list = [time_spent , stage_fear , social_event_attendance , going_out , drained , friend_circle , Post_frequency]
```

```python
feature_list
```

```python
def yesandno(string):
    value = string.strip().lower()
    if value.startswith('y'):
        return 1
    elif value.startswith('n'):
        return 0
    else:
        print("Just a small nudge 😊 — please enter Yes or No.")
        return "retry"

def preprocessing_the_user_input(l):
    # Extract numerical values first
    numeric_values = [l[0], l[2], l[3], l[5], l[6]]

    # Scale them all at once
    scaled_values = scaler.transform([numeric_values]).flatten().tolist()

    # Process Yes/No answers
    yn_values = [yesandno(l[1]), yesandno(l[4])]

    if "retry" in yn_values:
        print("One or more invalid Yes/No inputs. Try again.")
        return None

    # Combine scaled numeric values with yes/no values in the original order
    # Order: [scaled 0], yn 1, [scaled 2, 3], yn 4, [scaled 5, 6]
    processed = [
        scaled_values[0],     # Time_spent_Alone
        yn_values[0],         # Stage_fear
        scaled_values[1],     # Social_event_attendance
        scaled_values[2],     # Going_outside
        yn_values[1],         # Drained_after_socializing
        scaled_values[3],     # Friends_circle_size
        scaled_values[4],     # Post_frequency
    ]

    return processed
```
